```python
def get_unavailable_packages(error_file):
    unavailable_packages = []
    with open(error_file, 'r') as file:
        for line in file:
            if ' - ' in line:
                package = line.split(' - ')[1].strip()
                unavailable_packages.append(package)
    return unavailable_packages

def remove_packages(requirements_file, unavailable_packages):
    with open(requirements_file, 'r') as file:
        lines = file.readlines()

    with open(requirements_file, 'w') as file:
        for line in lines:
            if not any(package in line for package in unavailable_packages):
                print(line)
                file.write(line)
```

```python
# Replace 'error.txt' with the path to your error output file
unavailable = get_unavailable_packages('error.txt')
# Replace 'requirements.txt' with the path to your requirements file
remove_packages('requirements.txt', unavailable)
```
